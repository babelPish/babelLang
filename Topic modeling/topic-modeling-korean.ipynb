문자열에 대해 명사 추출
KoNLPy를 통해 꼬꼬마 형태소 분석기를 이용하여 한글 문자열로부터 명사를 얻을 수 있다.

```python
# -*- coding: utf-8 -*-
from konlpy.tag import Kkma
from konlpy.utils import pprint
import re
```

```python
from konlpy.corpus import kolaw
```

```python
doc = kolaw.open('constitution.txt').read()
```

```python
print(doc)
```

```python
def text_cleaning(text):
    text = re.sub('[^가-힝0-9a-zA-Z]', ' ', text)
    return text
```

```python
doc = text_cleaning(doc)
```

```python
print(doc)
```

```python
nouns = Kkma().nouns(doc)
```

```python
print(nouns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
vectorizer = TfidfVectorizer(lowercase=False)
```

```python
print(vectorizer)
```

```python
x_list = vectorizer.fit_transform(nouns)
```

```python
print(x_list)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda = LatentDirichletAllocation(n_components=3, learning_method = 'batch').fit(x_list)
```

```python
print(lda)
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
print(feature_names)
```

```python
keywords = []
```

```python
def top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        keywords = [" ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]])]
    return keywords
```

```python
keywords.append(top_words(lda, feature_names, 5))
```
