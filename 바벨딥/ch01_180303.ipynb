## 1-1.데이터 다운로드하기

 스크레이핑, 크롤링
* 스크레이핑 : 웹 사이트에 있는 특정 정보를 추출하는 기술
* 크롤링 : 프로그램이 웹 사이트를 정기적으로 돌며 정보를 추출하는 기술 / 크롤러, 스파이더




<ol>urllib : URL을 다루는 모듈을 모아놓은 라이브러리</ol>
<ol>urllib.request : 웹사이트에 있는 데이터에 접근하는 기능을 제공하는 모듈</ol>

<h3>urllib.request를 이용한 다운로드</h3>

```python
# 웹사이트에서 파일을 다운로드할 때는 urllib.request 모듈에 있는 urlretrieve()함수를 사용 - 직접 파일 다운 가능

# 라이브러리 읽어 들이기 --- (※1)
import urllib.request  # urllib 라이브러리에 내부에 있는 request 모듈을 가져오기

# 다운받을 URL과 이름 지정하기
url = "http://uta.pw/shodou/img/28/214.png"
savename = "test.png"

# 다운로드 --- (※2)
urllib.request.urlretrieve(url, savename)  #url이 가리키는 주소에 접근해서 해당 자원을 로컬에 저장해주는 기능
print("저장되었습니다...!")
```

<h3>urlopen()으로 파일에 저장하는 방법</h3>

```python
# request.urlretrieve() 함수를 이용해 파일에 곧바로 저장
# request.urlopen() 곧바로 파일이 저장되는 것이 아니라, 데이터를 파이썬 메모리 위에 올릴 수 있다.

import urllib.request

# URL과 저장 경로 지정하기
url = "http://uta.pw/shodou/img/28/214.png"
savename = "test1.png"

# 다운로드 --- (※1)
mem = urllib.request.urlopen(url).read()   # urlopen에서 url을 가져오고 read()로 호출
# http://hayeon9894.tistory.com/25


# 파일로 저장하기 --- (※2)
with open(savename, mode="wb") as f:   #쓰기모드, 바이너리 모드
    f.write(mem)
    print("저장되었습니다...!")
```

```python
# 텍스트파일 활용

import urllib.request

url = "http://news.naver.com/"

mem = urllib.request.urlopen(url).read()
#print(mem)
print(mem.decode("euc-kr"))
```

<h3>웹에서 데이터 추출하기</h3>

### GET 요청 기본

사용자가 URL을 브라우저 주소창에 작성하고 엔터를 누르면 원하는 웹페이지가 나온다. 사용자는 웹페이지를 보기위해 단순한 일을 한 것 이지만, 특정 웹페이지를 사용자 웹브라우저에게 보여주기 위해서는 내부적인 처리들이 있다. 그 내부적인 처리에서, 클라이언트가 서버에게 웹페이지를 보여달라고 말하는 것을 요청 이라 부른다.



#### ex>초콜릿 검색
https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%B4%88%EC%BD%9C%EB%A6%BF

    
웹 요청메서드

* 요청 방식 : GET, POST, PUT, DELETE....    # http://terms.naver.com/entry.nhn?docId=2271986&cid=51207&categoryId=51207


* 요청 대상 : https://search.naver.com  -> 이곳에 요청을 하게됨. 호스트 이름



* 추가적인 정보   
-경로 : /search.naver   
-데이터(요청매개변수,  GET방식에서는 '?'뒤) : <키>=<값>을 &로 연결  
?where=nexearch  
&sm=top_hty  
&fbm=1  
&ie=utf8  
&query=%EC%B4%88%EC%BD%9C%EB%A6%BF(인코딩) <- 초콜릿(디코딩)
 
    
* 인코딩이란 정보를 부호화/암호화시킨다, 디코딩은 그 부호화/암호화를 해제한다는 뜻

```python
import urllib.request
import urllib.parse
# urllib.parse 모듈은 url에 한글 검색어를 입력할 수 있도록 도와주는 모듈. 인코딩 해주는 모듈

#https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%B4%88%EC%BD%9C%EB%A6%BF

naver = "https://search.naver.com/search.naver" #대상
values = {
    "where":"nexearch",
    "sm": "top_hty",
    "fbm": "1",
    "ie": "utf8",
    "query": "초콜릿" # %EC%B4%88%EC%BD%9C%EB%A6%BF  
}   

params = urllib.parse.urlencode(values)   #인코딩  
# 파이썬으로 요청 전용 매개변수를 만들 때는 urllib.parse 모듈의 urlencode()함수를 사용해 매개변수를 url 인코딩 한다
# 매개변수에 한국어 등이 포함되어있다면 반드시 url 인코딩을 해야한다

print(naver)
print(params)

url = naver + "?" + params
print(url)
```

```python
data = urllib.request.urlopen(url).read()
#print(data)

text = data.decode("utf-8")    #euc-kr
print(text)
```

<h3>URL에 매개변수를 추가해 요청을 전송하는 방법</h3>
<ol>기상청의 RSS데이터 - 지정 지역번호에 해당하는 정보 제공</ol>
<ol>http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp</ol>

```python
# 경상북도 정보 출력 

import urllib.request
import urllib.parse
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 URL 인코딩합니다. --- (※1)
values = {
    'stnId': '143'
}
params = urllib.parse.urlencode(values)

# 요청 전용 URL을 생성합니다. --- (※2)
url = API + "?" + params
print("url=", url)

# 다운로드합니다. --- (※3)
data = urllib.request.urlopen(url).read()
#print(data)
text = data.decode("utf-8")
print(text)
```
